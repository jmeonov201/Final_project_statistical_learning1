## Arbol de decisión con sklearn

```python
from sklearn import tree
from sklearn.metrics import brier_score_loss
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.externals import joblib ## libreria para poder escribrir, guardar o exportar los modelos al disco duro
```

## Función para decisión tree

```python
def train_tree(XT, YT, XV, YV,PASO):
    tree_model = tree.DecisionTreeClassifier(max_depth=PASO)
    tree_model = tree_model.fit(XT, YT) ## a esto le deberia de hacer return si quiero usar el modelo para predecir 
    
    Y_Train_predic = tree_model.predict(XT)
    Error_Train = brier_score_loss(YT, Y_Train_predic)
    Accuracy_Train = accuracy_score(YT, Y_Train_predic, normalize = True)
    Precision_Train = precision_score(YT, Y_Train_predic) 
    Recall_Train = recall_score(YT, Y_Train_predic)
    F1_Train = f1_score(YT, Y_Train_predic)
    
    Y_Validation_predic = tree_model.predict(XV)
    Error_Validation = brier_score_loss(YV, Y_Validation_predic)
    Accuracy_Validation = accuracy_score(YV, Y_Validation_predic, normalize = True)
    Precision_Validation = precision_score(YV, Y_Validation_predic)
    Recall_Validation = recall_score(YV, Y_Validation_predic)
    F1_Validation = f1_score(YV, Y_Validation_predic)

    cadena_config = "Decisiontree""_max_depth="+str(PASO)
    ##return final_W, final_B, cadena_config
    
    
    print("Cadena:", cadena_config)
    
    print("Error de Entrenamiento:", Error_Train)
    print("Error de Vallidaciòn:", Error_Validation)
    
    print("Accuracy de Entrenamiento:", Accuracy_Train)
    print("Accuracy de Vallidaciòn:", Accuracy_Validation)
    
    print("Precision de Entrenamiento:", Precision_Train)
    print("Precision de Vallidaciòn:", Precision_Validation)
    
    print("Recall de Entrenamiento:", Recall_Train)
    print("Recall de Vallidaciòn:", Recall_Validation)
    
    print("f1 score de Entrenamiento:", F1_Train)
    print("f1 score de Vallidaciòn:", F1_Validation)
    
    return cadena_config,Error_Train,Accuracy_Train,Precision_Train,Recall_Train,F1_Train,Error_Validation,Accuracy_Validation,Precision_Validation,Recall_Validation,F1_Validation,tree_model 
```

```python
tcc1,te1,ta1,tp1,tr1,tf1,te1v,ta1v,tp1v,tr1v,tf1v,d_t_modelo1 = train_tree(X_train, Y_train, X_validation, Y_validation,None)
```

## Función para SVM

```python
def train_svm(XT, YT, XV, YV,c):
    svm_model = svm.SVC(C=c)
    svm_model = svm_model.fit(XT, YT) ## a esto le deberia de hacer return si quiero usar el modelo para predecir
    
    Y_Train_predic = svm_model.predict(XT)
    Error_Train = brier_score_loss(YT, Y_Train_predic)
    Accuracy_Train = accuracy_score(YT, Y_Train_predic, normalize = True)
    Precision_Train = precision_score(YT, Y_Train_predic) 
    Recall_Train = recall_score(YT, Y_Train_predic)
    F1_train = f1_score(YT, Y_Train_predic)
    
    Y_Validation_predic = svm_model.predict(XV)
    Error_Validation = brier_score_loss(YV, Y_Validation_predic)
    Accuracy_Validation = accuracy_score(YV, Y_Validation_predic, normalize = True)
    Precision_Validation = precision_score(YV, Y_Validation_predic)
    Recall_Validation = recall_score(YV, Y_Validation_predic)
    F1_Validation = f1_score(YV, Y_Validation_predic)
    
    cadena_config = "SVM""_C="+str(c)
    ##return final_W, final_B, cadena_config
    
    
    print("Cadena:", cadena_config)
    
    print("Error de Entrenamiento:", Error_Train)
    print("Error de Vallidaciòn:", Error_Validation)
    
    print("Accuracy de Entrenamiento:", Accuracy_Train)
    print("Accuracy de Vallidaciòn:", Accuracy_Validation)
    
    print("Precision de Entrenamiento:", Precision_Train)
    print("Precision de Vallidaciòn:", Precision_Validation)
    
    print("Recall de Entrenamiento:", Recall_Train)
    print("Recall de Vallidaciòn:", Recall_Validation)
    
    print("f1 score de Entrenamiento:", F1_train)
    print("f1 score de Vallidaciòn:", F1_Validation)
    
    ##cadena_config = "regLog""_epochs="+str(epochs)+"_lr="+str(LR)+"_Lasso="+str(L1)+"_Ledge="+str(L2)
    ##return final_W, final_B, cadena_config
    
    return cadena_config,Error_Train,Accuracy_Train,Precision_Train,Recall_Train,F1_train,Error_Validation,Accuracy_Validation,Precision_Validation,Recall_Validation,F1_Validation,svm_model
```

## batches e hiperparametros

```python
lr = 0.01
epochs = 20
batch_size = 32
batches = int(X_train.shape[0] / batch_size) ## extrae el numero de filas del dataset
```

## grafo 

```python
grafologistica = tf.Graph()
with grafologistica.as_default():
    
    with tf.name_scope("Declaring_placeholder"):    
        ## definiendo las X, 6 es el valor pues 6 son las x features luego del reshape 
        X = tf.placeholder(tf.float32, [None,5], name="X")
    
        ## definiendo las Y
        Y = tf.placeholder(tf.float32, [None,1], name="Y")
    
        ## estableciendo el learning rate placeholder
        ph_lr = tf.placeholder(dtype = tf.float64, name="LR")
        
        ## establenciend el valor de lamda para l1 y l2
        l1_g = tf.placeholder(dtype = tf.float32, name="L1")
        l2_g = tf.placeholder(dtype = tf.float32, name="L2")
        
    
    with tf.name_scope("Declaring_variables"):
        ## vector de parametros 
        W = tf.Variable(tf.zeros([5,1]), name = "W")
        B = tf.Variable(0, name = "b", dtype="float")
    
    with tf.name_scope("hipotesis"):   
        ## caculando los logits
        logits_g = tf.add(tf.matmul(X, W, name="hipotesis_producto_punto"), B, name="hipotesis_sumar_B")
    
        ## obteniendo las yhat pasando los logits por sigmoid
        ## esta sire para predecir
        yhat_g = tf.nn.sigmoid(logits_g, name="sigmoid")
    
    with tf.name_scope("funcion_costo"):
        ## calculando la fucion de costo con la entropia cruzada
        ## esta srive para entrenar
        cost_g = tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(logits = logits_g, labels = Y))+tf.multiply(tf.reduce_sum(tf.abs(W)),l1_g)+tf.multiply(tf.reduce_sum(tf.pow(W,2)),l2_g) 
        ##aca tengo que poner la regularizacion lasso o ridge
        ##aca tengo que poner la regularizacion lasso o ridge [lamda * tf.sum(tf.abs(w),vertical axis =1)]
    
    with tf.name_scope("Gradient_Des.Optimizer"):
        ## gradient descent para optimizar un nodo mas del grafo
        optimizacion = tf.train.GradientDescentOptimizer(ph_lr).minimize(cost_g)
               
    ## accuracy a pie o a mano 
    correct_prediction = tf.equal(tf.cast(tf.greater_equal(yhat_g,0.5),tf.float32), Y)
    accuracy = tf.reduce_mean(tf.cast(correct_prediction, tf.float32)) ##media de los aciertos me da el accuracy
    
    ## metricas con tensor flow
    yhat_g_cerouno = tf.cast(tf.greater_equal(yhat_g,0.5),tf.float32)
    
    
    precision_g = tf.metrics.precision(Y, yhat_g_cerouno)
    recall_g = tf.metrics.recall(Y, yhat_g_cerouno)
    #f1_g = tf.metrics.f1_score(Y, yhat_g_cerouno)
    
```

```python
def train_logistica(XT, YT, XV, YV, epochs=20, LR=0.1, L1=0, L2=0):
    with tf.Session(graph = grafologistica) as session:
        # incializando variables
        tf.global_variables_initializer().run()
        tf.local_variables_initializer().run() #esto se puso para que corriera el accuracy de tensorflow
        print('Initialized')
    
        for epoch in range(epochs):
            for i in range(batches):
                offset = i * epoch 
                x = XT[offset: offset + batch_size] ## slicing de filas
                y = YT[offset: offset + batch_size] ## slicing de filas
                _, final_W, final_B, train_accuracy, _yhatrain = session.run([optimizacion, W, B, accuracy, yhat_g_cerouno], 
                                                                  feed_dict={X:x, Y:y, ph_lr:LR, l1_g:L1, l2_g:L2})
                
                cost_s = session.run(cost_g, feed_dict={X:x, Y:y, ph_lr:LR, l1_g:L1, l2_g:L2})
                                                          
            if (epoch % 1 == 0):
                print("Iteración en entrenamiento:", epoch, " Costo en entranmiento:", cost_s)
        
        print("Accuracy de entrenamiento:", train_accuracy)
        #print("accuracy_t:", accuracy_t)
    
        ## usando el grafo con otros datos 
        errorvalida, accuracyvalida, _yhatvalida = session.run([cost_g, accuracy, yhat_g_cerouno], 
                                                               feed_dict={X:XV, Y:YV, l1_g:L1, l2_g:L2})
        print("Costo en validación:", errorvalida)
        print("Accuracy de validación:", accuracyvalida)
            
    #Cadena de configuraciòn
    cadena_config = "regLog""_epochs="+str(epochs)+"_lr="+str(LR)+"_Lasso="+str(L1)+"_Ledge="+str(L2)
    
    ## metricas en entrenamiento 
    Precision_Train = precision_score(y, _yhatrain) 
    Recall_Train = recall_score(y, _yhatrain)
    F1_Train = f1_score(y, _yhatrain)
    
    ## metricas en validacion
    Precision_Validation = precision_score(YV, _yhatvalida) 
    Recall_Validation = recall_score(YV, _yhatvalida)
    F1_Validation = f1_score(YV, _yhatvalida)
        
    
    return cadena_config,cost_s,train_accuracy,Precision_Train,Recall_Train,F1_Train,errorvalida,accuracyvalida,Precision_Validation,Recall_Validation,F1_Validation,final_W, final_B, 
```
